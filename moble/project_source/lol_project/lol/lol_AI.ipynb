```python
import tensorflow as tf
import numpy as np #배열 처리를 위해 사용
from pandas.io.parsers import read_csv #엑셀 데이터를 읽어오기 위해 필요
```

```python

#메모장으로 키면 , 으로 분리되어있기 때문 (csv파일이기 때문)
data = read_csv('../lol_data/total_4.csv', sep=',') # '를 토큰으로 데이터를 분리해서 담을수 있도록 
#print(data)

#정수형도 소수점형태로 받아와야한다.
xy = np.array(data, dtype=np.float32) #행렬 형태로 해당 데이터를 담을 수 있도록 함
```

```python
print(xy)
```

```python
#10개의 변인을 입력을 받습니다.
x_data = xy[:,0:-1] # 10가지 챔피언정보

#가격 값을 입력 받습니다.
y_data = xy[:,[-1]] # 글로벌 골드
```

```python
print(y_data)
```

```python
X = tf.placeholder(tf.float32, shape=[None,10], name="cham") #10개의 변인이 담길수 있게
Y = tf.placeholder(tf.float32, shape=[None,1], name="gold")

#배열 형식의 난수 만들기
W = tf.Variable(tf.random_normal([10,1]), name="weight") #가중치
b = tf.Variable(tf.random_normal([1]), name="bias") #bias 값
```

```python
#기본적으로 선형 회귀같은 경우는 행렬의 곱 연산을 이용해서 결과식을 세울수 있다.
#가설식 하이포테시스
# M행 K열 / K행 M열
# 행렬의 곱은 K열과 K행이 같아야 성립할 수 있다.
# 같아야 두 식이 이어질수가 있어서 곱셈이 가능하다.
# 앞에서는 행이 필요하고 뒤에서는 열이 필요하다
# 행렬의 곱셈은 곱셈의 결과의 해석이 중요하다.
# https://www.youtube.com/watch?v=qwEWO-cQwXQ

#학생들이 햄버거를 살때 낼 돈을 표라는 그래프로 나타낸것! (가격과, 살 갯수를 연결시켜 놓은 것)
hypothesis = tf.matmul(X,W, name="operation") +b 
```

```python
#비용함수
#우리가 세운 가설이 얼마나 정확한지 판단하는 기준
#데이터와 가설의 직선거리를 계산해서 판단함
cost = tf.reduce_mean(tf.square(hypothesis-Y), name="result")
```

```python
#최적화 함수
#학습의 보폭을 적절히 선택해야 안정적인 학습이 가능하다.
#0.000005
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.000005,name="opt") #학습률 정도(매우 중요)
train = optimizer.minimize(cost,name="tra") #훈련 설정
```

```python
#세션을 만들어야 학습을 시킬수 있음 (계산을 한다.)
sess = tf.Session()
#변수를 초기화해줘야 변수 대입이 된다.
sess.run(tf.global_variables_initializer())
```

```python
for step in range(1000001):
    cost_, hypo_, _ = sess.run([cost,hypothesis, train], feed_dict={X: x_data, Y: y_data})
    if step % 500 == 0:
        print("#", step, " 손실 비용: ", cost_)
        print("- 배추 가격: ",hypo_[0])#예측되고 있는 배추 가격 #첫번째 데이터에 대해서
```

```python
saver = tf.train.Saver()
#학습모델 이용에 있어서 최적화 등의 요소는 고려하지 않고 단순하게 세이버를 이용해서 학습된 모델을 추출하는방법
save_path = saver.save(sess,"./test1/test.saved.cpkt")
```

```python
#테스트 실험
X = tf.placeholder(tf.float32, shape=[None,10]) #4개의 변인이 담길수 있게
Y = tf.placeholder(tf.float32, shape=[None,1]) 
W = tf.Variable(tf.random_normal([10,1]), name="weight") #가중치
b = tf.Variable(tf.random_normal([1]), name="bias") #bias 값

hypothesis = tf.matmul(X,W) +b 

#저장된것을 가져오기 위한 객체 초기화
saver = tf.train.Saver()
model= tf.global_variables_initializer()

#사용자로부터 10가지 변수를 입력 받음
top= float(input('top : '))
jg = float(input('jg : '))
mid = float(input('mid : '))
ad = float(input('ad : '))
sup = float(input('sup : '))

e_top= float(input('e_top: '))
e_jg = float(input('e_jg: '))
e_mid = float(input('e_mid: '))
e_ad = float(input('e_ad: '))
e_sup = float(input('e_sup: '))

with tf.Session() as sess:
    sess.run(model)
    
    save_path="./saved.cpkt"
    saver.restore(sess,save_path) #해당 세션에 저장된 학습 모델을 불러올수 있도록 함
    
    #기존에 학습된 데이터가 2차원 배열이기 때문에 맞추어 줘야 한다.
    data = ((top, jg, mid, ad, sup, e_top, e_jg, e_mid, e_ad, e_sup), )
    
    #사용자가 입력한 데이터를 토대로 행렬로 초기화를 해주고
    arr = np.array(data, dtype=np.float32)
    
    x_data = arr[0:10]
    #입력한 데이터를 X에 담아서 모델을 돌려보았다
    dict = sess.run(hypothesis, feed_dict={X: x_data})
    
    #데이터를 하나만 넣었기 때문에 출력은 하나만 하면 된다.
    print(dict[0])
```
